```
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,np.pi,1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

La linea de comandos x = np.arange(0,4*np.pi,0.1) nos genera un espacio de valores dentro del intervalo deseado, para el ejemplo mencionado podemos ver como la reprecentacion grafica comienza en 0, el cual es el primer valor del intervalo y finaliza en 4*PI (3,14), con un paso o incremento de 0,1 en x para poder realizar la grafica.
Como podemos ver en el segundo codigo al modificar los valores preestablecidos podemos realizar la grafica desde un nuevo valor de x que en este caso es de -10 y finaliza en PI y al modificar el ultimo parametro de la misma podemos ver como la grafica deja de tener la suabidad de curvas que presentaba ya que aumentamos el paso a 1
